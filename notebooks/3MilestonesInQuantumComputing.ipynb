### 3.1 Feynman's Vision

In 1981, Richard Feynman, proposed a new kind of computer upon contemplating the implications of quantum mechanics: "You can simulate this with a quantum system, with quantum computer elements. It’s not a Turing machine, but a machine of a different kind".He observed that classical machines hit an exponential wall. For a system of $R$ particles, there exists a function called the amplitude consisting of the particles at position $x_1, x_2, \ldots, x_R$ at time $t$. This simply had too many variables, and it could not be simulated with a normal computer. In the 1980s, Yuri Manin and Paul Benioff also showed that a superposed quantum system has a vastly bigger state‑space than a classical system. Evidently, classical space has $N$ discrete levels, however superposed quantum systems have $c^N$ discrete levels. 

Formally, John Bell showed that no classical "hidden variable" theory could reproduce the predictions of quantum mechanics. However, even before this was widely accepted, Feynman proposed that quantum mechanics could be used to simulate itself.

### 3.2 Early Foundations (1980-1995)
- In 1985, **David Deutsch** formalised the concept of a **universal quantum computer**, which could simulate any quantum system. He also introduced the idea of quantum gates and circuits, laying the groundwork for quantum algorithms.
- In 1992, **Umesh Vazirani** showed that by leveraging superposition, you could **define tasks** that are provably hard for any classical algorithm but **feasible on a quantum computer**.
- In 1994, **Peter Shor** developed a quantum algorithm for factoring large integers and finding discrete logarithms, demonstrating that quantum computers could solve certain problems exponentially faster than classical computers. This was a major breakthrough, showing the potential of quantum computing for cryptography.



### 3.3 Challenges in Quantum Computing (1996)

1. **Quantum Decoherence**:

   In a perfect scenario, we want to maintain the delicate quantum states needed for computation so we can manipulate them manually. A large reason quantum computing is challenging is due to **quantum decoherence**, which is the loss of quantum information due to interactions with the environment. This can cause errors and noise in quantum computations, making it very difficult to maintain delicate quantum states. 

   Very quickly after Shor's breakthrough, early papers by Unruh and others showed that error accumulation and thermal noise impose strict time limits on reliable quantum operations, hence quantum algorithms could not outperform classical algorithms in practice. This led to the development of **quantum error correction** techniques, which allow quantum computers to detect and correct errors without measuring the quantum state directly.

   - In 1996, the first **quantum error correction codes** were proposed by Shor, Steane, and others. Essentially, we can achieve this by encoding one “logical” qubit into many “physical” qubits, creating redundancy that allows for the detection and correction of errors without ever measuring (and collapsing) the single qubit itself.
   - In late 1996, Aharonov and Ben-Or proved there existed a **fault tolerance threshold**. If physical gate error rates are below a certain threshold, then arbitrarily long quantum computations can be performed reliably (but at a cost of many physical qubits). This was a major breakthrough, showing that quantum error correction could make quantum computing feasible in practice.

   The question then became: "How will we scale up to quantum computing systems that can solve hard problems?". Researchers envisioned using error correction and engineering to build large-scale quantum computers, but this was a significant task. Error correction requires many physical qubits to encode a single logical qubit, and the overhead of this was quite substantial.


2. **Applications of Quantum Computing**:

   Another challenge was quantifying quantum computing's industry uses. Intuitively, the best application of quantum computing would be to solve problems that are easy with quantum computers but hard for classical computers.

   In the 1990s, David Deutsch and Richard Jozsa proposed a problem that quantum computers could solve in a single query, yet any deterministic classical algorithm would need exponentially many queries in the worst case. This led to the **Deutsch–Jozsa algorithm**, which is regarded as the first quantum algorithm.

   The Deutsch-Jozsa problem was the **oracle problem**, which consists of a black box function. Essentially, you have an unknown function (a "black box") that takes an input and returns either 0 or 1, and you want to find out which input gives a specific output. We are told that the function is either constant ($f(x)=0$ for all $x$) or balanced (returns 0 for half the inputs and 1 for the other half). The challenge is to determine which type of function it is with as few queries as possible (ideally one query).

   1. **Quantum Circuit Setup**  
      - Start with $n$ “input” qubits in $\lvert0\rangle^{\otimes n}$ and one “output” qubit in $\lvert1\rangle$.  
      - Apply Hadamard gates to all $n+1$ qubits, creating a superposition of every input and putting the output qubit into $\frac{\lvert0\rangle - \lvert1\rangle}{\sqrt2}$.

   3. **Oracle Query**  
      - The oracle is a unitary $U_f$ that maps  
      $$ \lvert x\rangle\lvert y\rangle \;\mapsto\; \lvert x\rangle \lvert y \oplus f(x)\rangle,
      $$
      where $x$ is the $n$-bit input register and $y$ is the output qubit.  
      - Acting on our superposition, it imprints the function values as phase flips on the input register.
   4. **Hadamard Transform**
      - Apply Hadamard gates again to the **input** register only.  
      - If $f$ is constant, **all** amplitudes for $\lvert0\rangle^{\otimes n}$ constructively interfere to 1, and every other basis state cancels to 0.  
      - If $f$ is balanced, the amplitude for $\lvert0\rangle^{\otimes n}$ becomes 0.
   5. **Measurement & Quantum Advantage**
      - Measure the input register. If the outcome is $\lvert0 \ldots 0 \rangle $, we conclude that $f$ is constant. Otherwise, $f$ is balanced.
      - This quantum algorithm requires only **one query** to the oracle, demonstrating a quantum advantage over classical algorithms, which would require at least $2^{n-1} + 1$ queries in the worst case.

In the 2000s, researchers began to develop physical systems that could implement quantum computers. These systems were based on various technologies:
- **Atomic Ions**: a single, electrically charged atom (ion) can be trapped and manipulated using electromagnetic fields. We use two of its energy levels as qubit states, for example a ground state $\lvert g \rangle$ as $\lvert0\rangle$ and a long-lived excited state $\lvert e \rangle$ as $\lvert1\rangle$. Laser pulses can be used to manipulate the ion's state, allowing for quantum gates to be implemented.
- **Superconducting Circuits**: individual photons carry qubit information in properties like polarization, path (which of two waveguides a photon travels), or time-bin encoding (early vs late arrival). Linear optics quantum computing uses beamsplitters and phase shifters to manipulate these photons, allowing for quantum gates to be implemented.
- **Neutral atoms**: electrically neutral atoms, e.g. rubidium or strontium, are trapped one‑by‑one in tightly focused laser beams (“optical tweezers”). Two long‑lived hyperfine levels of the atom encode the $ \lvert0\rangle$ and $\lvert1\rangle$ qubit states. Global or site‑selective laser or microwave fields drive single‑qubit rotations which can be used to implement quantum gates.

Each platform trades off different strengths like coherence time, gate speed, connectivity, and manufacturability, but all share the same abstract concept of two-level quantum systems to encode the $ \lvert0\rangle$ and $\lvert1\rangle$ qubit states.

### 3.5 Modern Quantum Computing (Present)
**Noisy Intermediate-Scale Quantum (NISQ)** devices are the current state of quantum hardware. Instead of using many extra physical qubits to encode a single logical qubit, NISQ devices apply quantum error correction **after** running a noisy circuit to partially undo errors. For example, running the same circuit at several noise levels and extrapolating back to “zero noise". However, the amount of classical work (and number of repeated runs) you need grows exponentially if you want to push error rates arbitrarily close to zero, and this is only practical for very small circuits.

A more holistic approach, as mentioned earlier, is to encode each logical qubit into many physical qubits using a quantum code, but this may require dozens to thousands of physical qubits for every logical qubit (depending on how low final error is to be)!


**Fault-Tolerant Quantum Computing (FTQC)** will require millions of high-quality qubits with gate error rates far below what we currently have to run complex algorithms, which is a significant engineering challenge. This might take decades to achieve, but it is the ultimate goal of quantum computing research.

Nevertheless, breakthroughs in quantum hardware and algorithms continue to be made:
- **Quantum Supremacy**: In 2019, Google claimed to have achieved quantum supremacy with their Sycamore processor, performing a specific calculation that would take classical supercomputers thousands of years in just 200 seconds. 
- **Harvard, MIT and QuEra**: In 2023, researchers from Harvard, MIT and QuEra sampled 48 logical qubits using only a 280-qubit device, representing a significant step in reducing the overhead required for quantum error correction.
- **Willow Processor**: In 2024, Google's 105-qubit Willow processor demonstrated a significant improvement in quantum error correction, achieving a logical qubit with a record low error rate.


Even today, there exist many open applications in quantum computing that are being actively researched.
- **Quantum Chemistry**: 
    - In 1929, Dirac stated "Quantum mechanics has a problem: the equations are much too complicated to be soluble.”
    - Yet, heuristic classical algorithms are often very successful at predicting molecular properties, and quantum computers are expected to be able to do this more efficiently.
- **Quantum Dynamics**:
    - Real-time evolution of many-body quantum systems too hard classically. Quantum computers can (theoretically) implement Trotter-Suzuki methods to simulate time evolution of quantum systems.
